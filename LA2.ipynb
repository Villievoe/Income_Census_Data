```python
import pandas as pd
import numpy as np
import matplotlib as plt

df = pd.read_csv("/home/vinutna/Documents/Data Science Class/adult_data.csv")
ageTime = df.groupby("occupation").agg({"Hrs-per-week": "mean"})
plotAgeTime = ageTime.plot(color = "#ab1234")
```

```python
import matplotlib.pyplot as pplt

eduGain = df.groupby("education").agg({"finalweight": "max"})
plotEduGain = eduGain.plot.pie(subplots = "True",legend=False)

centre_circle = pplt.Circle((0,0),0.70,fc='white')
fig = pplt.gcf()
fig.gca().add_artist(centre_circle)

pplt.tight_layout()
pplt.show()
```

```python
regIncome = df.groupby("Native-country").agg({"finalweight": "mean"})
plotRegIncome = regIncome.plot(kind = "bar", width = 0.4, facecolor = np.random.rand(3,))
```

```python
relWeight = df.groupby("relationship").agg({"Capital-gain" : "max"})
plotRelWeight = relWeight.plot(kind = "bar", facecolor = np.random.rand(3,))
```

```python
import seaborn as sns
import pandas as pd
import matplotlib as plt
import numpy as np

df = pd.read_csv("/home/vinutna/Documents/Data Science Class/adult_data.csv")
sns.set_style("whitegrid")
sns.stripplot(x="race",y="finalweight",data=df,palette=sns.mpl_palette("Set1",5))
sns.despine()
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="Income-range",y="Hrs-per-week",data=df)
sns.despine()
```

```python
sns.set_style("whitegrid")
sns.violinplot(x=df['sex'],y=df['age'],palette=sns.mpl_palette("Set3",2))
sns.despine()
```

```python
sns.pointplot(x=df['Capital-loss'],y=df['workclass'],palette="bright")
sns.despine()
```

```python
sns.lineplot(x=df['Capital-loss'],y=df['age'],color="violet",data=df)
sns.despine()
```
